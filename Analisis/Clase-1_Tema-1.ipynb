**Distribuciones discretas**

```python
# bernouli
# bernoulli de scipy.stats para generar valores con la distribución de Bernoulli.
# matplotlib.pyplot para graficar resultados.
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

Distribución de **Bernoulli**

```python
# Probabilidad de autenticación exitosa
p = 0.92
# Simular 100 autenticaciones
resultados = bernoulli.rvs(p, size=100)
# Contar resultados
exitos = sum(resultados)
fallos = len(resultados) - exitos
print(f"Éxitos: {exitos}, Fallos: {fallos}")

# Se suman los valores 1 para contar los éxitos.
# Los fallos se calculan como el complemento (100 - éxitos).
# Se imprime un resumen de cuántos intentos funcionaron y cuántos fallaron.
```

**Distribución Normal**

```python
from scipy.stats import norm
# Parámetros del problema
mu= 250     # media
sigma = 30  # desviación estándar
x = 300     # umbral deseado
# Calcular probabilidad de que la respuesta sea > 300 ms
probabilidad = 1 - norm.cdf(x, mu, sigma)
print(f"Probabilidad de respuesta > 300 ms: {probabilidad:.4%}")
```

```python
import numpy as np
from scipy import stats
# Datos de muestra
tiempos = [12, 11, 13, 12, 14, 10, 11, 15, 13, 12]
n = len(tiempos)
media_muestra = np.mean(tiempos)
desv_std_muestra = np.std(tiempos, ddof=1) # ddof=1 para muestras

# Intervalo de confianza del 95%
confianza = 0.95
grados_libertad = n-1
error_estandar = desv_std_muestra / np.sqrt(n)

t_critico = stats.t.ppf((1 + confianza) / 2, df=grados_libertad)
margen_error = t_critico * error_estandar

lim_inf = media_muestra - margen_error
lim_sup = media_muestra + margen_error

print(f"Intervalo de confianza 95%: [{lim_inf:.2f}, {lim_sup:.2f}] días")
```
