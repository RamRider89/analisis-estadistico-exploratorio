# 🐍 Aplicación del Algoritmo KNN para Clasificación

Para este ejemplo, utilizaremos un conjunto de datos ficticio de diagnóstico médico (común en problemas de clasificación) para predecir si un paciente tiene una condición (Clase 1) o no (Clase 0) basado en mediciones.

1. Preparación y Estandarización de Datos
El escalado de características es fundamental para KNN, ya que las distancias se basan en la magnitud de las variables.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np

# 1. Crear un dataset de ejemplo (simulado)
data = {
    'Feature_A': np.random.rand(100) * 100,
    'Feature_B': np.random.rand(100) * 5,
    'Feature_C': np.random.rand(100) * 0.1,
    'Target': np.random.randint(0, 2, 100) # 0 o 1
}
df = pd.DataFrame(data)

X = df[['Feature_A', 'Feature_B', 'Feature_C']]
y = df['Target']

# 2. División en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Estandarización (Preprocesamiento Obligatorio para KNN)
# StandardScaler transforma los datos para que tengan media 0 y desviación estándar 1
scaler = StandardScaler()

# Ajustar (fit) el escalador SÓLO en los datos de entrenamiento para evitar fuga de datos
X_train_scaled = scaler.fit_transform(X_train)

# Aplicar (transform) a los conjuntos de entrenamiento y prueba
X_test_scaled = scaler.transform(X_test)

print("Datos listos y escalados.")
```

# 2. Entrenamiento y Evaluación del Modelo KNN
Aplicaremos el modelo y evaluaremos su rendimiento utilizando las métricas clave.

```python
# 1. Aplicación del Modelo KNN
K = 5 # Seleccionamos 5 vecinos (valor común para empezar)
knn = KNeighborsClassifier(n_neighbors=K)

# Entrenar el modelo con los datos ESCALADOS
knn.fit(X_train_scaled, y_train)

# 2. Realizar predicciones
y_pred = knn.predict(X_test_scaled)

# 3. Evaluación del rendimiento
print("\n" + "="*50)
print(f"Resultados de KNN con K = {K}")
print("="*50)

# Exactitud (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Exactitud (Accuracy): {accuracy:.4f}")

# Matriz de Confusión
print("\nMatriz de Confusión:\n", confusion_matrix(y_test, y_pred))

# Reporte de Clasificación (Precision, Recall, F1-score)
print("\nReporte de Clasificación (por clase):\n", classification_report(y_test, y_pred))
```

| Métrica               | Definición                                                               | Interpretación                                                                                                                                |
|-----------------------|--------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------------------------------------|
| Exactitud (Accuracy)  | Proporción total de clasificaciones correctas.                           | Si es alta (ej., 0.93), indica que la mayoría de las predicciones son correctas en general. Pero es engañosa con datos desbalanceados.        |
| Matriz de Confusión   | Muestra el recuento de aciertos y errores por clase (TP, TN, FP, FN).    | Fundamental. Permite ver si el modelo confunde la Clase 0 con la Clase 1, revelando desequilibrios en el rendimiento por clase.               |
| Recall (Sensibilidad) | Capacidad del modelo para encontrar todas las instancias positivas.      | Si la Clase 1 (condición) es vital, un alto Recall es clave, ya que significa pocos Falsos Negativos (no diagnosticar a un paciente enfermo). |
| Precisión             | Proporción de identificaciones positivas que fueron realmente correctas. | Si el costo de un Falso Positivo (diagnosticar erróneamente a un paciente sano) es alto, esta métrica debe ser alta.                          |
| F1-Score              | Media armónica de Precisión y Recall.                                    | El mejor indicador individual del rendimiento equilibrado de una clase.                                                                       |


# 3. Análisis del Impacto de K y Preprocesamiento
- A. Impacto del Parámetro K (Número de Vecinos)La elección de $K$ es un balance entre el sesgo y la varianza:

| **$K$ Pequeño (ej., 1 o 3)** | El modelo es flexible y se ajusta a los datos locales. | **Alto _Overfitting_** (sobreajuste) al ruido y a los valores atípicos. Alto Sesgo. |
| **$K$ Grande (ej., 25 o 50)** | El modelo es suave y generaliza mejor. | **Alto _Underfitting_** (subajuste). Las fronteras de decisión se vuelven muy amplias, ignorando patrones importantes. |
