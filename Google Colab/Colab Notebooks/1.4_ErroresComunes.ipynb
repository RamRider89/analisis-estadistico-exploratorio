# 1.4 Errores comunes en visualización de datos
## Propósito del cuaderno
Este cuaderno documenta errores típicos que distorsionan la interpretación, a veces sin querer… y a veces queriendo.

Sofía advierte que una mala visualización puede llevar a decisiones equivocadas en comités.

### Errores frecuentes:
- Eje truncado sin avisar → exagera cambios pequeños.
- Gráficos 3D "bonitos" → sacrifican legibilidad.
- Chartjunk → decoraciones, sombras, íconos, stickers que no aportan información.
- Paletas engañosas → colores que sugieren alarma donde no hay alarma.
- Doble eje mal usado → mezcla escalas distintas y confunde causa/efecto.

---
## 1.4.1 Práctica guiada: Eje truncado
Vamos a simular ingresos mensuales con un crecimiento moderado y luego generar:

1. Una versión manipulada (eje Y truncado cerca del valor mínimo), que hace ver el crecimiento como dramático.
2. Una versión honesta (eje Y inicia en 0), que muestra el crecimiento real.

Al final construiremos una tabla de reflexión ética.


```python
# Paso 1. Simular datos
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Mes':['Ene','Feb','Mar','Abr','May','Jun'],
    'Ingresos':[95,98,105,110,115,118]  # millones simulados
})

df
```

```python
# Paso 2. Versión manipulada (eje truncado)
plt.figure(figsize=(8,4))
plt.bar(df['Mes'], df['Ingresos'], color='tomato')
plt.ylim(90, 120)  # <-- eje truncado muy arriba de 0
plt.title('Versión alarmista: parece un salto enorme')
plt.ylabel('Ingresos (millones)')
plt.show()

# Paso 3. Versión honesta (escala completa)
plt.figure(figsize=(8,4))
plt.bar(df['Mes'], df['Ingresos'], color='steelblue')
plt.ylim(0, 120)  # <-- escala iniciando en 0
plt.title('Versión honesta: crecimiento gradual y estable')
plt.ylabel('Ingresos (millones)')
plt.show()
```

```python
# Paso 4. Tabla de reflexión ética
reflexion = pd.DataFrame({
    'Aspecto evaluado':[ 'Magnitud percibida del cambio',
                         'Sensación emocional que genera',
                         'Posible decisión del lector' ],
    'Gráfico eje truncado':[ 'El salto parece enorme',
                              'Sensación de urgencia / crisis',
                              'Reacción rápida, tal vez exagerada' ],
    'Gráfico eje honesto':[ 'El cambio es progresivo',
                             'Sensación de control / estabilidad',
                             'Acción de seguimiento y ajuste, no pánico' ]
})
reflexion
```
