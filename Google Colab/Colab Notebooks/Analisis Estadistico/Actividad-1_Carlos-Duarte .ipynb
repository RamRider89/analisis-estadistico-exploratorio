# 🔍 Análisis estadístico descriptivo

## ⚡ Frecuencia y consumo energético

En esta sección, se calcularan las medidas de tendencia central y dispersión para las columnas Freq (MHz) y TDP (W).

Se realizará el cálculo de las siguientes medidas estadísticas para Freq y TDP:
 - Promedio, mediana, desviación estándar, mínimo, máximo y rango.


```python
# estadísticas para la columna Freq
freq_promedio = df[_columna_freq].mean()
freq_mediana = df[_columna_freq].median()
freq_desviacion = df[_columna_freq].std()
freq_min = df[_columna_freq].min()
freq_max = df[_columna_freq].max()
freq_rango = freq_max - freq_min

# estadísticas para la columna TDP
tdp_promedio = df[_columna_tdp].mean()
tdp_mediana = df[_columna_tdp].median()
tdp_desviacion = df[_columna_tdp].std()
tdp_min = df[_columna_tdp].min()
tdp_max = df[_columna_tdp].max()
tdp_rango = tdp_max - tdp_min
```

### 📑 Resultado del análisis estadístico para Freq y TDP

Se presentan y organizan los resultados obtenidos en tablas y/o visualizaciones para identificar patrones o comparaciones entre ambas variables.

```python
# creando un nuevo dataframe con los resultados
cols_stds = ['Promedio', 'Mediana', 'Desviación estándar', 'Mínimo', 'Máximo', 'Rango']
data_estadistica = {
    'Promedio': [freq_promedio, tdp_promedio],
    'Mediana': [freq_mediana, tdp_mediana],
    'Desviación estándar': [freq_desviacion, tdp_desviacion],
    'Mínimo': [freq_min, tdp_min],
    'Máximo': [freq_max, tdp_max],
    'Rango': [freq_rango, tdp_rango]
}
df2 = pd.DataFrame(data_estadistica, columns=cols_stds)
df2.index = [_columna_freq, _columna_tdp]

print("\n")
print("\033[1m --- Análisis de frecuencia y consumo energético --- \033[0m")
print("_" * 80)
df2.round(decimals=2)
#print(tabulate(df2, headers='keys', tablefmt='psql'))
```

## 🧭 Estadísticas comparativas por fabricante

En esta sección se agrupan los datos por la columna **Vendor** para generar estadísticas comparativas para cada fabricante.

Se utilizará el método .agg() de **Pandas** para calcular múltiples estadísticas a la vez para las columnas Freq y TDP.

```python
# realizamos una búsqueda de los diferentes fabricantes
fabricantes = df[_columna_vendor].unique()
# arreglo con las diferentes funciones estadísticas
sts_arr = ['mean', 'median', 'std', 'max', 'min']

# se agrupan los datos por la columna 'Vendor' y se calculan las estadísticas descriptivas
stats_fabricante = df.groupby(_columna_vendor).agg({
    _columna_product: 'count',
    _columna_freq: sts_arr,
    _columna_tdp: sts_arr,
    _columna_transistors: sts_arr,
})

print("\n")
print("\033[1m --- Estadísticas comparativas por fabricante --- \033[0m")
print("_" * 110)
stats_fabricante.round(decimals=2)
```

### 📊 Gráfica de cantidad de productos por fabricante

```python
# configurando un grafico de barras comparativo entre marcas
# creacion de un dataset para el total de productos
total_prods = pd.DataFrame(stats_fabricante[_columna_product]['count'].reset_index())
x_data = total_prods[_columna_vendor]
y_data = total_prods['count']

# creando la grafica de barras
plt.figure(figsize=(10, 5))
ax = sns.barplot(
    data=total_prods,
    x=x_data,
    y=y_data,
    hue=x_data,
    palette=_paleta_tecmi
)

# añadiendo etiquetas a cada barra
for i in range(len(x_data)):
    x = x_data[i]
    y = y_data[i]
    # texto de la etiqueta formateado
    label = f"{y:.2f}"
    ax.annotate(
        label,
        (x, y),
        textcoords="offset points",
        xytext=(0, 1),
        ha='center',
        color=_color_secondary,
        fontsize=fontsize_marker)

# definimos los titulos
plt.title('Cantidad de productos analizados', fontsize=fontsize_title)
plt.xlabel('Fabricante', fontsize=fontsize_label)
plt.ylabel('Cantidad', fontsize=fontsize_label)
plt.show()
```

A continuación se genera un gráfico de barras múltiples para comparar el rendimiento promedio, los transistores y el consumo de energía entre los distintos fabricantes.

```python
# configurando un grafico de barras comparativo entre marcas
# creacion de un dataset para cada caracteristica, freq, tdp, trans
prom_frecuencias = pd.DataFrame(stats_fabricante[_columna_freq]['mean'].reset_index())
prom_tdps = pd.DataFrame(stats_fabricante[_columna_tdp]['mean'].reset_index())
prom_transistors = pd.DataFrame(stats_fabricante[_columna_transistors]['mean'].reset_index())

# se insertan las columnas adicionales para identificar el tipo de valor
prom_frecuencias.insert(1, "Type", _columna_freq)
prom_tdps.insert(1, "Type", _columna_tdp)
prom_transistors.insert(1, "Type", _columna_transistors)

# unimos a los tres datasets en uno solo para procesarlo en la grafica
todos_promedio = pd.concat([prom_frecuencias, prom_tdps, prom_transistors], ignore_index=True)
todos_promedio = todos_promedio.sort_values(by='mean')
#print(todos_promedio)

# grafica tipo catplot
ax = sns.catplot(
    todos_promedio,
    kind="bar",
    x="Type",
    y="mean",
    col="Vendor",
    height=5,
    aspect=0.8,
    hue="Type",
    palette=_paleta_boostrap_min,
    legend=True
)

# titulos
ax.set_xlabels("Característica")
ax.set_ylabels("Valor promedio")
plt.show()

```

### 🗂 Prueba t de Student para AMD vs. Intel

*La prueba t de Student nos ayudará a determinar si las medias de las frecuencias de los procesadores de AMD e Intel son significativamente diferentes entre sí.*

```python
# filtrando datos de frecuancias de los fabricantes AMD e INTEL
amd_freq = df[df[_columna_vendor] == _AMD][_columna_freq]
intel_freq = df[df[_columna_vendor] == _INTEL][_columna_freq]

# prueba t de Student
t_stat, p_valor_t = stats.ttest_ind(amd_freq, intel_freq, equal_var=False)

# columnas para presentar informe
_col_estadistico = 'Estadístico t'
_col_p = 'P-valor'

print("\033[1m --- Resultados de la prueba t de Student (" + _AMD + " vs. " + _INTEL + ") --- \033[0m")
print('-' *60)
dfA = pd.DataFrame({_col_estadistico: [t_stat], _col_p: [p_valor_t]})
print(tabulate(dfA, headers=[_col_estadistico, _col_p], tablefmt="grid", floatfmt=".4f"))
```

**Interpretación**:
- El **p-valor** es **menor** que **0.05**, por lo que podemos **rechazar** la **hipótesis nula** y concluir que hay una diferencia estadísticamente significativa.


## 🅱 Hipótesis B

Se definen las siguientes hipótesis:
- Hipótesis Nula (H₀): Las medias de frecuencia de los grupos (AMD, Intel, NVIDIA, ATI) son iguales.
- Hipótesis Alternativa (H₁): Al menos una de las medias de frecuencia es significativamente diferente de las demás.

### 🗂 Prueba ANOVA para tres o más fabricantes

```python
# filtrando los datos para todos los fabricantes
amd_freq = df[df[_columna_vendor] == _AMD][_columna_freq].dropna()
intel_freq = df[df[_columna_vendor] == _INTEL][_columna_freq].dropna()
nvidia_freq = df[df[_columna_vendor] == _NVIDIA][_columna_freq].dropna()
ati_freq = df[df[_columna_vendor] == _ATI][_columna_freq].dropna()
otros_freq = df[df[_columna_vendor] == _Other][_columna_freq].dropna()

# confirmando que los grupos incluyan información y no esten vacios
if len(amd_freq) > 0 and len(intel_freq) > 0 and len(nvidia_freq) > 0 and len(ati_freq) > 0:
    # ejecutando prueba ANOVA
    f_stat, p_valor_anova = stats.f_oneway(amd_freq, intel_freq, nvidia_freq, ati_freq)

    # columnas para presentar informe
    _col_estadistico_f = 'Estadístico F'

    print("\033[1m --- Resultados de la prueba ANOVA (" + _AMD + " vs. " + _INTEL + " vs. " + _NVIDIA + " vs. " + _ATI + ") --- \033[0m")
    print('-' *70)
    dfB = pd.DataFrame({_col_estadistico_f: [f_stat], _col_p: [p_valor_anova]})
    print(tabulate(dfB, headers=[_col_estadistico_f, _col_p], tablefmt="grid", floatfmt=".4f"))
else:
    print("No se pueden realizar la prueba ANOVA, uno o más grupos están vacíos.")
```

**Interpretación:**
- El **p-valor** de la prueba *ANOVA* es **menor** que **0.05**, por lo que podemos **rechazar** la **hipótesis nula** y concluir que hay una diferencia significativa en la frecuencia promedio entre al menos uno de los grupos.

## 📖 Conclusión de las dos hipotesis propuestas


### 🎯 Conclusión basada en las hipótesis

- Hipótesis Nula (H₀): Las medias de frecuencia de los grupos (AMD, Intel, NVIDIA, ATI) son iguales.
- Hipótesis Alternativa (H₁): Al menos una de las medias de frecuencia es significativamente diferente de las demás.


- Dado que el **p-valor** (0.0000) es mucho menor que el nivel de significancia (0.05), en ambas pruebas, debemos rechazar la hipótesis nula H₀. Esto significa que hay suficiente evidencia estadística para aceptar la hipótesis alternativa H₁.
 - **Prueba t de Student**: Se rechaza H₀. **La conclusión es que sí existe una diferencia significativa en la frecuencia promedio entre los procesadores de AMD e Intel.**
 - **Prueba ANOVA**: Se rechaza H₀. **La conclusión es que sí existe una diferencia significativa en la frecuencia promedio entre al menos uno de los fabricantes AMD, Intel, NVIDIA o ATI.**

```python
# limpiando los datos de la fecha y transistores
df_clean = df.dropna(subset=[_columna_release_date, _columna_transistors, _columna_tdp]).copy()
df_clean[_columna_release_year] = pd.to_datetime(df_clean[_columna_release_date]).dt.year

# -------------------------------------------------------------------------- #
# por grupo
# columnas adicionales para el tratamiento de los datos
col_promedio_tdp = 'promedio_tdp'
col_promedio_transistors = 'promedio_transistors'
col_promedio_transistors_scaled = 'promedio_transistors_scaled'

# calculando el historico promedio anual de transistores y TDP
datos_anuales = df_clean.groupby(_columna_release_year).agg(
    promedio_transistors=(_columna_transistors, 'mean'),
    promedio_tdp=(_columna_tdp, 'mean')
).reset_index()

# se escalan los datos de transistores para que se visualicen mejor en el mismo gráfico que TDP
max_tdp = datos_anuales[col_promedio_tdp].max()
max_transistors = datos_anuales[col_promedio_transistors].max()
datos_anuales[col_promedio_transistors_scaled] = (datos_anuales[col_promedio_transistors] / max_transistors) * max_tdp

# -------------------------------------------------------------------------- #
# individual
max_tdp_o = df_clean[_columna_tdp].max()
max_transistors_o = df_clean[_columna_transistors].max()
df_clean[col_promedio_transistors_scaled] = (df_clean[_columna_transistors] / max_transistors_o) * max_tdp_o
# -------------------------------------------------------------------------- #

# configuración de la gráfica
fig, ax1 = plt.subplots(figsize=(15, 7))

# -------------------------------------------------------------------------- #
# configurando el primer eje Y para el TDP -> ax1
sns.lineplot(
    data=df_clean,
    x=_columna_release_year,
    y=_columna_tdp,
    ax=ax1,
    color=_color_danger,
    label='TDP promedio (W)',
    linewidth=2.5,
    markers=False,
    dashes=False,
    legend="full",
    alpha=0.8
)

# -------------------------------------------------------------------------- #
# configurando el segundo eje Y para los transistores -> ax2
ax2 = ax1.twinx()
sns.lineplot(
    data=datos_anuales,
    x=_columna_release_year,
    y=col_promedio_transistors_scaled,
    ax=ax2,
    color=_color_warning,
    label='Transistores promedio (millones)',
    linewidth=2.0,
    markers=False,
    dashes=False,
    legend="full",
    alpha=0.8,
    marker='o'
)

# -------------------------------------------------------------------------- #
# añadiendo etiquetas a cada punto -> ax1
x_data_ax1 = datos_anuales[_columna_release_year]
y_data_ax1 = datos_anuales[col_promedio_tdp]
for i in range(len(x_data_ax1)):
    x = x_data_ax1[i]
    y = y_data_ax1[i]
    # texto de la etiqueta formateado
    label = f"{y:.2f}"
    ax1.annotate(
        label,
        (x, y),
        textcoords="offset points",
        xytext=(0, 10),
        ha='center',
        color=_color_secondary,
        fontsize=fontsize_marker)

# -------------------------------------------------------------------------- #
# añadiendo etiquetas a cada punto -> ax2
x_data_ax2 = datos_anuales[_columna_release_year]
y_data_ax2 = datos_anuales[col_promedio_transistors_scaled]
for i in range(len(x_data_ax2)):
    x = x_data_ax2[i]
    y = y_data_ax2[i]
    # texto de la etiqueta formateado
    label = f"{y:.2f}"
    ax2.annotate(
        label,
        (x, y),
        textcoords="offset points",
        xytext=(0, 10),
        ha='center',
        color=_color_secondary,
        fontsize=fontsize_marker)



# -------------------------------------------------------------------------- #
# titulos ax1
ax1.set_title('La revolución de la eficiencia: Transistores vs. consumo energético', fontsize=fontsize_title, pad=20)
ax1.set_xlabel('Año de lanzamiento', fontsize=fontsize_label)
ax1.set_ylabel('TDP promedio (W)', fontsize=fontsize_label)
ax1.tick_params(axis='y')
plt.legend(loc='upper left', title='Leyenda')
# titulos ax2
ax2.set_ylabel('Transistores promedio (millones)', fontsize=fontsize_label)
ax2.tick_params(axis='y')
# show
ax1.grid(False)
ax2.grid(False)
plt.legend(loc='upper center')
fig.tight_layout()
plt.show()

```
