### 2.2: Análisis de varianza (ANOVA) y prueba chi-cuadrado

```python
from scipy import stats
#Carga de datos
soporte = [4.2, 4.0, 4.1, 4.3, 4.2]
facturacion = [3.8, 3.7, 3.9, 3.8, 3.6]
consultoria = [4.5, 4.6, 4.4, 4.7, 4.6]
#Cálculo de la anova oneway y del p-valor
f_stat, p_valor = stats.f_oneway(soporte, facturacion, consultoria)
print("F:", round(f_stat, 3), "| p-valor:", round(p_valor, 4))
```

```python
import numpy as np
from scipy.stats import chi2_contingency
# Construcción de la tabla
tabla = np.array([
    [35, 10, 5],    # Laptop
    [30, 20, 10],   # Escritorio
    [20, 25, 15]    # Móvil
])
# Prueba chi-cuadrado
chi2, p, dof, esperados = chi2_contingency(tabla)
print("Chi²:", round(chi2, 3), "| p-valor:", round(p, 4))
```

```python
# Gráfica 2: Detalle del BoxPlot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset
df = sns.load_dataset("tips")

# Crear el boxplot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x=df["tip"], color="lightblue")
plt.title("Distribución de propinas con elementos etiquetados", fontsize=14)

# Calcular valores estadísticos
q1 = df["tip"].quantile(0.25)
q3 = df["tip"].quantile(0.75)
iqr = q3 - q1
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

# Añadir anotaciones
plt.axvline(q1, color="orange", linestyle="--", label="Q1 (25%)")
plt.axvline(q3, color="orange", linestyle="--", label="Q3 (75%)")
plt.axvline(lower_whisker, color="green", linestyle="--", label="Bigote inferior")
plt.axvline(upper_whisker, color="green", linestyle="--", label="Bigote superior")

# Outliers
outliers = df[(df["tip"] < lower_whisker) | (df["tip"] > upper_whisker)]["tip"]
for outlier in outliers:
    plt.text(outlier, 0.1, f"{outlier:.2f}", rotation=45, color="red", fontsize=9)

plt.xlabel("Tip")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips") #Importar el dataset Tips de Seaborn

# Calcular Q1 y Q3 para la columna 'tip' (propinas)
Q1 = df['tip'].quantile(0.25)
Q3 = df['tip'].quantile(0.75)

# Calcular el rango intercuartil (IQR)
IQR = Q3 - Q1

# Definir límites inferior y superior para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar valores que estén fuera del rango aceptable
outliers = df[(df['tip'] < limite_inferior) | (df['tip'] > limite_superior)]

# Mostrar los valores atípicos encontrados
print(outliers)
```

```python
# Gráfica 3: Propinas antes y después de filtrar los outliers.

# Sin outliers
df_filtrado = df[(df['tip'] >= limite_inferior) & (df['tip'] <= limite_superior)]

sns.histplot(df['tip'], color="red", label="Original", kde=True)
sns.histplot(df_filtrado['tip'], color="blue", label="Filtrado", kde=True)
plt.legend()
plt.title("Propinas antes y después del filtrado de outliers")
plt.show()
```

### 3.3 Gestión de datos faltantes, imputación y técnicas de limpieza

```python
# Cargar y visualizar el dataset
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
df.info()


```

```python
# Gráfica 4: Valores faltantes Titanic

# Visualizar los datos faltantes

import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu")
plt.title("Mapa de valores faltantes en Titanic")
plt.show()
```

```python
# Ver cuántos datos faltan:
print(df.isnull().sum().loc[lambda x: x > 0].sort_values(ascending=False))
```

```python
#Imputar edad con media

media_edad = df['age'].mean()
df['age'] = df['age'].fillna(media_edad)

# Ver cuántos datos faltan:
print(df.isnull().sum().loc[lambda x: x > 0].sort_values(ascending=False))
```

```python
#Imputar 'embarked' con la moda (valor más frecuente)

moda_embarked = df['embarked'].mode()[0]
df['embarked'] = df['embarked'].fillna(moda_embarked)

# Ver cuántos datos faltan:
print(df.isnull().sum().loc[lambda x: x > 0].sort_values(ascending=False))
```

```python
# Eliminar la columna 'deck' por tener más del 70% de NaN

df = df.drop(columns=['deck'])

# Ver cuántos datos faltan:
print(df.isnull().sum().loc[lambda x: x > 0].sort_values(ascending=False))
```

```python
# Valores únicos de embark_town
print(df['embark_town'].unique())

#registros con NaN de embarket y embarket_town
print(df[df['embark_town'].isnull()][['embarked', 'embark_town']])

```

```python
# Sustituir NaN con valores conocidos de acuerdo con una condición:

# Crear el mapeo automáticamente desde datos no nulos
mapa = df.dropna(subset=['embark_town']).drop_duplicates(subset=['embarked'])[['embarked', 'embark_town']].set_index('embarked')['embark_town']

# Aplicar imputación directamente
df.loc[df['embark_town'].isnull(), 'embark_town'] = df.loc[df['embark_town'].isnull(), 'embarked'].map(mapa)

# Ver cuántos datos faltan:
print(df.isnull().sum().loc[lambda x: x > 0].sort_values(ascending=False))

```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Cargar y preparar datos
df = sns.load_dataset("penguins").dropna()
cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

# Crear versiones escaladas
df_original = df[cols].copy()
df_minmax = df[cols].copy()
df_zscore = df[cols].copy()

# Escalamiento Min-Max
scaler_minmax = MinMaxScaler()
df_minmax[cols] = scaler_minmax.fit_transform(df_minmax)

# Escalamiento Z-Score
scaler_zscore = StandardScaler()
df_zscore[cols] = scaler_zscore.fit_transform(df_zscore)

# Añadir etiquetas
df_original['Escalado'] = 'Original'
df_minmax['Escalado'] = 'Min-Max'
df_zscore['Escalado'] = 'Z-Score'

# Función para guardar gráficos
def save_plot(df, title, filename):
    df_melted = pd.melt(df, id_vars='Escalado', var_name='Variable', value_name='Valor')
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df_melted, x='Variable', y='Valor', palette='Set2')
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filename)
    plt.close()

# Guardar imágenes
save_plot(df_original, "Boxplot - Escala Original", "boxplot_original.png")
save_plot(df_minmax, "Boxplot - Min-Max Scaling", "boxplot_minmax.png")
save_plot(df_zscore, "Boxplot - Z-Score Scaling", "boxplot_zscore.png")
```

### 6.5 Impacto de PCA en modelos

```python
# Paso 1. Preparar los datos

import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Cargar y limpiar dataset
penguins = sns.load_dataset("penguins").dropna()

# Seleccionar variables numéricas predictoras
X = penguins[["bill_length_mm", "bill_depth_mm", "flipper_length_mm"]]
y = penguins["body_mass_g"]

# Estandarizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
# Paso 2. Modelo original sin PCA

modelo_original = LinearRegression()
modelo_original.fit(X_train, y_train)
y_pred_orig = modelo_original.predict(X_test)

r2_orig = r2_score(y_test, y_pred_orig)
mse_orig = mean_squared_error(y_test, y_pred_orig)

print(f"R² original: {r2_orig:.3f}")
print(f"MSE original: {mse_orig:.2f}")
```

```python
# Paso 3. Aplicar PCA y reentrenar

# Aplicar PCA (2 componentes)
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Modelo con PCA
modelo_pca = LinearRegression()
modelo_pca.fit(X_train_pca, y_train)
y_pred_pca = modelo_pca.predict(X_test_pca)

r2_pca = r2_score(y_test, y_pred_pca)
mse_pca = mean_squared_error(y_test, y_pred_pca)

print(f"R² con PCA: {r2_pca:.3f}")
print(f"MSE con PCA: {mse_pca:.2f}")
```

### 8.2 Modelado de series temporales en entornos empresariales

```python
#Carga y visualización inicial

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset desde una URL
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
df.rename(columns={"Passengers": "Passengers"}, inplace=True)

# Visualización de la serie temporal
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["Passengers"], label="Pasajeros mensuales", color="blue")
plt.title("SkyRoute Airlines - Número de pasajeros mensuales (1949–1960)")
plt.xlabel("Fecha")
plt.ylabel("Número de pasajeros (en miles)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Descomposición de la serie

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Usamos la serie ya cargada en un DataFrame llamado df con columna 'pasajeros'
# Asegurarnos que el índice sea de tipo datetime y tenga frecuencia mensual
df.index = pd.to_datetime(df.index)
df = df.asfreq('MS')  # MS: Month Start

# Aplicar descomposición estacional aditiva
result = seasonal_decompose(df['Passengers'], model='multiplicative')

# Graficar resultados
result.plot()
plt.suptitle("Descomposición Estacional de la Serie de Pasajeros", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Modelo SARIMA: (p=1,d=1,q=1)x(P=1,D=1,Q=1,s=12)

from statsmodels.tsa.statespace.sarimax import SARIMAX

# Modelo SARIMA: (p=1,d=1,q=1)x(P=1,D=1,Q=1,s=12)
sarima_model = SARIMAX(
    df['Passengers'],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_results = sarima_model.fit(disp=False)
print(sarima_results.summary())
```

```python
# pronóstico de los próximos 12 valores

from statsmodels.tsa.statespace.sarimax import SARIMAX
import pandas as pd

# Paso 1: Dividir los datos
n_periods = 12
train_data = df['Passengers'][:-n_periods]
test_data = df['Passengers'][-n_periods:]

# Paso 2: Ajustar el modelo SARIMA
model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12),
                enforce_stationarity=False, enforce_invertibility=False)
results = model.fit(disp=False)

# Paso 3: Pronosticar los próximos 12 valores
forecast = results.get_forecast(steps=n_periods)
predicted_mean = forecast.predicted_mean

# Paso 4: Crear la tabla comparativa
comparison_df = pd.DataFrame({
    'Fecha': test_data.index,
    'Real': test_data.values,
    'Predicción SARIMA': predicted_mean.values,
    'Error Absoluto': abs(test_data.values - predicted_mean.values)
})

# Mostrar la tabla
print(comparison_df)
```
