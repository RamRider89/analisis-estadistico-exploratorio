NumPy

```python
import numpy as np
# datos de tiempo de atencion
tiempos = np.array([3.2, 4.1, 2.9, 3.8, 3.5])

# calculos
media = np.mean(tiempos)
desviacion = np.std(tiempos, ddof=1)

# resultados
print(f"Media: {media:.2f} min")
print(f"Desv. estandar: {desviacion:.2f} min")
# NumPy es 100x más rápido al evitar iteraciones en Python y paralelizar operaciones.
```

```python
suma = 0
for t in tiempos:
    suma += t
media_python = suma / len(tiempos)
print(f"media_python: {media_python:.2f} min")
```

***anova***

```python
from scipy import stats
# datos
soporte = [4.2, 4.0, 4.1, 4.3, 4.2]
facturacton = [3.8, 3.7, 3.9, 3.8, 6]
consuLtoria = [4.5, 4.6, 4.4, 4.7, 6]

# calculo de anova

f_stat, p_valor = stats.f_oneway(soporte, facturacton, consuLtoria)

print("F:", round(f_stat, 3), "| p-valor:", round(p_valor, 4))

if(p_valor > 0.05):
    print("se concluye que existen diferencias significativas en la satisfacción promedio de los clientes según el tipo de servicio recibido")
```

Prueba Chi-cuadrado

```python
import numpy as np
from scipy.stats import chi2_contingency

# tabla
tabla = np.array([
    [35, 10, 5],
    [30, 20, 10],
    [20, 25, 15]
])

# prueba
chi2, p, dof, esperados = chi2_contingency(tabla)

print("Chi2:", round(chi2, 3), "| p-valor:", round(p, 4))


if(p < 0.05):
    print("Rechazamos la hipótesis nula")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos simulados
data = {
    'horas_soporte': [1, 2, 2, 3, 4, 5, 5, 6, 7, 8],
    'satisfaccion': [5.0, 4.8, 4.7, 4.5, 4.0, 3.8, 3.5, 3.2, 2.8, 2.5]
}

df = pd.DataFrame(data)
sns.set(style="whitegrid")

# Diagrama de dispersión con línea de regresión
sns.lmplot(data=df, x='horas_soporte', y='satisfaccion', ci=95, line_kws={'color': 'red'})

plt.title("Horas de soporte vs Satisfacción del cliente")
plt.xlabel("Horas de soporte técnico")
plt.ylabel("Nivel de satisfacción (1 a 5)")
plt.show()
```
