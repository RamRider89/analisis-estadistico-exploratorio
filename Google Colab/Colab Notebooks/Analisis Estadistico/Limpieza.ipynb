# Manual Básico de Limpieza de Datos 🛠️
Este manual te guiará a través de las fases iniciales de la limpieza de datos, desde el diagnóstico hasta la aplicación de técnicas básicas. Usaremos la librería pandas como nuestra herramienta principal.

# Paso 1: Diagnóstico Inicial del Dataset 🩺
Antes de limpiar, debemos entender el estado de nuestros datos. Este primer paso nos ayuda a identificar los problemas más obvios como valores faltantes o tipos de datos incorrectos.

Script de Python:

```python
import pandas as pd
import seaborn as sns

# Cargar el dataset
# df = pd.read_csv('nombre_de_tu_archivo.csv')
df = sns.load_dataset('penguins')

# Mostrar las primeras 5 filas para una vista rápida
print("Vista previa del DataFrame:")
print(df.head())

# Obtener un resumen de la información: tipos de datos y valores nulos
print("\nInformación del DataFrame:")
print(df.info())

# Contar la cantidad de valores nulos por columna
print("\nConteo de valores nulos por columna:")
print(df.isnull().sum())
```

# Paso 2: Eliminación de Duplicados 👯
Los registros duplicados pueden sesgar cualquier análisis. Es una de las primeras tareas de limpieza que debes realizar.

Script de Python:

```python
# Contar el número de filas duplicadas antes de eliminarlas
duplicados = df.duplicated().sum()
print(f"\nNúmero de registros duplicados encontrados: {duplicados}")

# Eliminar las filas duplicadas
df_clean = df.drop_duplicates()

# Verificar que los duplicados se han eliminado
print(f"Número de registros después de eliminar duplicados: {df_clean.shape[0]}")
```

# Opción B: Eliminación de filas con muchos valores nulos

Si una fila tiene la mayoría de sus valores nulos, puede ser mejor eliminarla por completo.

```python
# Eliminar filas donde al menos 3 columnas tienen valores nulos
df_clean_final = df_clean.dropna(thresh=3)
print(f"\nRegistros restantes después de eliminar filas con demasiados nulos: {df_clean_final.shape[0]}")
```

# Paso 5: Conversión de Tipos de Datos 🔄
A menudo, las columnas que deberían ser numéricas o de fecha se cargan como texto. Es vital convertirlas para que sean útiles en el análisis.

Script de Python:

```python
# Convertir una columna a tipo numérico (float)
df_clean_final['columna_numerica'] = pd.to_numeric(df_clean_final['columna_numerica'], errors='coerce')
# El parámetro 'errors'='coerce' convierte los valores que no son números a NaN

# Convertir una columna a tipo de fecha (datetime)
df_clean_final['fecha'] = pd.to_datetime(df_clean_final['fecha'])
print("\nTipos de datos actualizados:")
print(df_clean_final.info())
```
