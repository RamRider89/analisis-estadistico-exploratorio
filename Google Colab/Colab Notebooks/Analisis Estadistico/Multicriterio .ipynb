```python
import pandas as pd
import numpy as np
```

```python
# definir los criterios y sus pesos
criterios = ['Costo', 'Calidad', 'Tiempo de entrega']
pesos = [0.40, 0.35, 0.25]

# definir los datos de los proveedores
datos = {
    'Proveedor': ['A', 'B', 'C'],
    'Costo': [10000, 12000, 11000],
    'Calidad': [8, 7, 9],
    'Tiempo de entrega': [5, 3, 4],
}

# crear un datafram
df = pd.DataFrame(datos)
df
```

```python
# normalizamos los datos
df['Costo'] = df['Costo'].max() / df['Costo']
df['Calidad'] = df['Calidad'] / df['Calidad'].max()
df['Tiempo de entrega'] = df['Tiempo de entrega'].max() / df['Tiempo de entrega']
df
```

```python
# calcilo del valor esperado de cada proveedor
df['Puntuacion total'] = (df['Costo']  * pesos[0]) + (df['Calidad']  * pesos[1]) + (df['Tiempo de entrega']  * pesos[2])
df
```

```python
# seleccionar el mejor proveedor
mejor_proveedor = df.loc[df['Puntuacion total'].idxmax()]
mejor_proveedor
```
